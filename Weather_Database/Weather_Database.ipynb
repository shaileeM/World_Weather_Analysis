```python
import pandas as pd
from config import weather_api_key
import requests
import matplotlib.pyplot as plt
import numpy as np
from citipy import citipy
```

```python
#Create a new set of 2,000 random latitudes and longitudes
lats=np.random.uniform(-90,90, size=2000)
lngs=np.random.uniform(-180,180,size=2000)
lats_lngs=zip(lats,lngs)
```

```python
#Add the zipped object into a list
coordinates=list(lats_lngs)
```

```python
#Find cities from coordinates, Get the nearest city using the citipy module.
cities=[]

for coordinate in coordinates:
    city=citipy.nearest_city(coordinate[0],coordinate[1]).city_name
    
    if city not in cities:
        cities.append(city)
        
len(cities)
```

```python
#Perform an API call with the OpenWeatherMap
city_data=[] #list for all weather data
base_url="http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key

for i,city in enumerate(cities):
    #print(city)
    city_url= base_url+"&q="+city.replace(" ","+")
    #print(city_url)
    try:
        city_response=requests.get(city_url).json()
        #print(city_response)
        city_lat=city_response['coord']['lat']
        #print(city_lat)
        city_country=city_response['sys']['country']
        #print(city_country)
        city_lon=city_response['coord']['lon']
        #print(city_lon)
        city_humidity=city_response['main']['humidity']
        #print(city_humidity)
        city_cloudliness=city_response['clouds']['all']
        #print(city_cloudliness)
        city_max_temp = city_response["main"]["temp_max"]
        #print(city_cloudliness)
        city_wind_speed=city_response['wind']['speed']
        #print(city_wind_speed)
        city_weather_description=city_response['weather'][0]['description']
        #Append into df
        city_data.append({"City": city,
                          "Country":city_country,
                          "Lat": city_lat,
                          "Lng": city_lon,
                          "Max Temp":city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_cloudliness,
                          "Wind Speed": city_wind_speed,
                          "Current Description":city_weather_description})
    except:
        print("City not found. Skipping...")
        pass
```

```python
print(city_data[1])
```

```python
# Convert the array of dictionaries to a Pandas DataFrame.
city_data_df = pd.DataFrame(city_data)
city_data_df.head(10)
```

```python
# Create the output file (CSV).
output_data_file = "WeatherPy_Database.csv"
# Export the City_Data into a CSV.
city_data_df.to_csv(output_data_file, index_label="City_ID")
```
